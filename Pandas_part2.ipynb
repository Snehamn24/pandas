# Sorting - Generally the sorting is done through the sort_values()

```python
#sorting the values by one column
sort_names = df.sort_values(by="Name")
sort_names.head(10)
```

```python
#sorting tehe values in descending order
sort_names_des = df.sort_values(by="Name",ascending = False)
sort_names_des.head(5)
```

```python
#converting the Birth column to the datatime column
df['Birth'] = pd.to_datetime(df['Birth'],errors='coerce')
df['Birth']
```

```python
#sort by the DOB
sort_dob = df.sort_values(by="Birth")
sort_dob.head(5)
```

```python
#sorting of 2 columns
#here if two person have a same name then it sorts using the Birth column otherwise it sort the columns using the 1st column i.e Name
sort_name_dob = df.sort_values(['Name','Birth'])
sort_name_dob.head(5)
```

```python
#in the below example it sorts the name in ascending order if it finds any duplicate names then it sorts the name in the descending order of the dob
sort_name_dob_desc = df.sort_values(['Name','Birth'],ascending=[True,False])
sort_name_dob_desc.head(5)
```
